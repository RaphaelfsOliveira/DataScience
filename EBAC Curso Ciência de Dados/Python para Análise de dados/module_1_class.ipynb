**exercício final:**

```python
svv_19 = 153.98
sqv_19 = 3

svv_20 = 337.01
sqv_20 = 7

svv_23 = 295.33
sqv_23 = 5

tkt_19 = svv_19 / sqv_19
tkt_20 = svv_20 / sqv_20
tkt_23 = svv_23 / sqv_23
print(f'''
Ticket médio por dia: 
Dia 19: {"{:.2f}".format(tkt_19)}
Dia 20: {"{:.2f}".format(tkt_20)}
Dia 23: {"{:.2f}".format(tkt_23)}
''')

tkt_total = sum([svv_19, svv_20, svv_23])
sqv_total = sum([sqv_19, sqv_20, sqv_23])

tkt_now = tkt_total / sqv_total
tkt_sum = sum([tkt_19, tkt_20, tkt_23]) / 3

print(f'''
Ticket médio atual: 
Média três dias: {"{:.2f}".format(tkt_sum)}
Média totais: {"{:.2f}".format(tkt_now)}
''')
```

### **2. Tipos Texto (string)**

```python
first_name = 'Raphael'
last_name = 'Oliveira'

hello = 'Meu nome é '+ first_name +' '+ last_name +'.'
print(hello)

hello = f'Meu nome é {first_name} {last_name}.'
print(hello)
```

```python
email = 'raphael@email.com'

print(len(email))

# for i, c in enumerate(email):
#   print(f'{i}: {c}')

print(email.upper())

print(email.find('email'))

print(email.replace('raphael', 'rp'))
```

```python
latlon = '-22.005; -47.89'

if latlon.find(';') != -1:
  lat, lon = latlon.split(';')
  
  print(f'''
    Lat: {'{:.3f}'.format(float(lat.strip()))}
    Lon: {'{:.3f}'.format(float(lon.strip()))}
  ''')
```

### **3. Tipos Boleanos**

```python
user = 'raphael'
password = '1234'
true, false = True, False

print(true, false, type(true), sep=' ', end='\n\n')
```

```python
saldo = 200
saque = 50
pode_sacar = saque <= saldo

print(pode_sacar)
```

```python
user_set = 'raphael'
password_set = '124'

user_check = user == user_set
password_check = password == password_set
access = user_check & password_check

print(user_check, password_check)
print(access)
```
