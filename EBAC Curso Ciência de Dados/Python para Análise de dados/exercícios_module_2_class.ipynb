## 1 Listas

> Criei uma lista chamada filmes com o nome dos 10 primeiros filmes mais bem avaliados no
site no IMDB. Imprima o resultado.

```python
movies = [
    'The Shawshank Redemption (1994)',
    'The Godfather (1972)',
    'The Godfather: Part II (1974)',
    'The Dark Knight (2008)',
    '12 Angry Men (1957)',
    "Schindler's List (1993)",
    'The Lord of the Rings: The Return of the King (2003)',
    'Pulp Fiction (1994)',
    'The Good, the Bad and the Ugly (1966)',
    'The Lord of the Rings: The Fellowship of the Ring (2001)',
]

def print_movies(movies):  
  for i, movie in enumerate(movies):
    print(f'Rank {i+1}: {movie}')

print_movies(movies)
```

> Simule a movimentação do ranking. Utilize os métodos insert e pop para trocar a posição
do primeiro e do segundo filme da lista. Imprima o resultado.

```python
first_movie = movies.pop(0)
movies.insert(1, first_movie)

print_movies(movies)
```

## 2 Conjuntos

> Aconteceu um erro no seu ranking. Simule a duplicação dos três últimos filmes da lista.
Imprima o resultado.

```python
movies.append(movies[7])
movies.append(movies[8])
movies.append(movies[9])

print_movies(movies)
```

> Utiliza a conversão set e list para remover os valores duplicados. Imprima o resultado. 

```python
movies_new = list(set(movies))

for movie in movies_new:
  print(f'{movie}')
```

## 3 Dicionários

> Repita os exercícios da parte 1 (listas). Os elementos da lista filmes devem ser dicionários
no seguinte formato: {'nome': <nome-do-filme>, 'ano': <ano do filme>},
'sinopse': <sinopse do filme>} .

```python
movies_list = []

for movie in movies:
  name = movie[:-6].strip()
  
  init_year = movie.find('(')
  final_year = movie.find(')')
  year = movie[init_year: final_year + 1]
  year = int(year.replace('(', '').replace(')', ''))

  movie_dict = {
      'nome': name,
      'ano': year,
      'sinopse': '...sadsadasdasd...',
  }

  movies_list.append(movie_dict)

for movie in movies_list:
  print(movie)

for movie in movies_list:
  print(f'''--
  Nome: {movie.get('nome')}
  Ano: {movie.get('ano')}
  Sinopse: {movie.get('sinopse')}'''
  )
```
