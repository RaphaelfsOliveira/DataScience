## Exercício Final

```python
saldo_inicial = 1000
transactions = []
```

```python
transactions.clear()

transactions.append(234)
transactions.append(-798.58)
transactions.append(472.12)
transactions.append(-10.91)

print(transactions)
```

```python
saldo_transactions = sum(transactions)
print(f'Saldo Inicial: {saldo_inicial}')
print(f'Saldo Transações: {saldo_transactions}')
print(f'Saldo Final: {saldo_inicial - abs(saldo_transactions)}')
```

# Conjuntos (set, tuple)

```python
norte_europa = {'reino unido', 'suecia', 'russia', 'noruega', 'dinamarca'}
escandinavia = {'suecia', 'noruega', 'dinamarca'}

norte_europa_nao_escandinavia = norte_europa - escandinavia
escandinavia_nao_norte_europa = escandinavia - norte_europa

print(norte_europa)
print(escandinavia)
print(norte_europa_nao_escandinavia)
print(escandinavia_nao_norte_europa)
```

# Dicionários (dict)

```python
brasil = {
    'capital': 'Brasília',
    'idioma': 'Português',
    'populacao': 210
}
print(
    brasil,
    type(brasil)
)
```

```python
credito = {
    '123': 750,
    '789': 980
}
```

```python
score_123 = credito['123']
score_789 = credito['789']

print(
    score_123,
    score_789
)
```

```python
credito.update({'123': 435})
print(credito)
```

```python
credito['456'] = 235
print(credito)
```

## Métodos

```python
artigo = dict(
    titulo='Modulo 02 | test',
    corpo='Topicos: .',
    total_char=1530
)

print(artigo)

artigo.update(dict(total_char=7850))

print(artigo)
```

```python
total_char = artigo.get('total_char')

print(artigo, total_char, sep='\n')
```

```python
chaves = artigo.keys()
print(chaves)
print(list(chaves))
```

```python
values = artigo.values()
print(values)
print(list(values))
```
