## if / else / elif

```python
true = 'Verdadeiro'
false = 'Falso'

if True:
  print(true)
else:
  print(false)
```

```python
cod_seg = '010'
cod_seg_cadastro = '010x'

psw = '7783'
psw_cad = '7783x'

can_pay = cod_seg == cod_seg_cadastro
psw_correct = psw == psw_cad
print(can_pay)
```

```python
if can_pay:
  print('Pagamento Efetuado')
else:
  print('Erro: Código de segurança inválido')
```

```python
if can_pay and psw_correct:
  print('Pagamento Efetuado')
else:
  print('Erro: Código de segurança inválido')
```

```python
if (cod_seg != cod_seg_cadastro) | (psw != psw_cad):
  print('Erro: Código de segurança inválido')
else:
  print('Pagamento Efetuado')
```

```python
if can_pay and psw_correct:
  print('Pagamento Efetuado')

elif not can_pay and psw_correct:
  print('Código de segurança inválido')

elif not psw_correct and can_pay:
  print('Senha inválida')

else:
  print('Erro: Código de segurança e senha inválidos')
```

```python
try:
  value = price / people
  print(value)
except ZeroDivisionError:
  print(f'Numero de pessoas invalido, o numero de pessoas tem que ser maior que zero, numero de pessoas: {people}')
```

```python
try:
  year = years[3]
  print(year)
except:
  print(f'Lista de anos menor que o valor escolhido, tem que ser de 0 a {len(years) -1}')
```

```python
try:
  year = years[3]
  print(year)
except Exception as error:
  print(f'Descrição da exceção: {error}')
  print(f'Tipo da exceção: {type(error)}')
  print(f'Lista de anos menor que o valor escolhido: {len(years)}')
```

```python
try:
  year = years[3]
  print(year)
except IndexError:
  print(f'Lista de anos menor que o valor escolhido, tem que ser de 0 a {len(years) -1}')
except Exception as error:
  print(f'Descrição da exceção: {error}')
  print(f'Tipo da exceção: {type(error)}')
  print(f'Lista de anos menor que o valor escolhido: {len(years)}')
```

```python
name = 'Raphael'
age = 30

try: 
  apr = name +' '+ age
  print(apr)

except TypeError:
  age = str(age)

except Exception as error:
  print(f'{error}')

finally:
  print('Second chance')
  apr = name +' '+ age
  print(apr)
```

## for / in

```python
for i, r in enumerate(range(6)):
  print(i, r)
```

```python
soma = 0

for v in range(0, 10):
  soma += v
  print(soma)

print(soma)
```

```python
for mult in range(2, 10, 3):
  print(mult)

print('-----')

for mult in range(10):
  if mult % 2 == 0:
    print(mult, '-> multiplo de dois')
  else:
    print(mult)
```

```python
fruits = ['maça', 'banana', 'laranja', 'uva', 'manga']

for fruit in fruits:
  print(fruit)
```

```python
credit = {
  '123': 750,
  '456': 812,
  '789': 980
}

for key, value in credit.items():
  print(f'Documento: {key}, score: {value}', end='\n\n')
```

```python
for key in credit.keys():
  print(f'chave: {key}')

print('-----')

for value in credit.values():
  print(f'valor: {value}')
```

```python
for i in range(100000):
  print(i)
  if i == 3:
    break
```

```python
numbers = range(10)

for num in numbers:
  if num % 2 == 0:
    print(f'O numero: {num} é par')
  else:
    print(f'O numero: {num} é impar')
  
  if num == 6:
    break
```
