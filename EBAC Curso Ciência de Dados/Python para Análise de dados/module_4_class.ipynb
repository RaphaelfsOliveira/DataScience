## Escrita

### Modos de arquivos
* r: leitura
* w: escrita (sobrescreve o arquivo original)
* a: abrir o arquivo para acrescentar


```python
with open(file='ages.csv', mode='w', encoding='utf8') as f:
  line = 'idade' + '\n'
  f.write(line)

  for age in ages:
    line = str(age) + '\n'
    f.write(line)
```

```python
content = None

with open(file='./ages.csv', mode='r', encoding='utf8') as f:
  content = f.read()

print(content)
```

```python
with open(file='ages.csv', mode='a', encoding='utf8') as f:
  for age in ages:
    line = str(age + 100) + '\n'
    f.write(line)
```

## Funções

```python
def to_upper(text: str) -> str:
  return text.upper()
```

```python
name = 'Raphael Oliveira'
print(name)

print(to_upper(name))
```

```python
def extract_user_prov(email: str) -> (str, str):
  email = email.split('@')
  user = email[0]
  prov = email[1]

  return user, prov
```

```python
email = 'raphael@email.com'
user, prov = extract_user_prov(email=email)

print(user)
print(prov)
```

```python
def write_file(file_path: str, header: str, content: list) -> bool:
  try:
    with open(file=file_path, mode='w', encoding='utf8') as fp:
      line = header + '\n'
      fp.write(line)

      for c in content:
        line = str(c) + '\n'
        fp.write(line)
  
  except Exception as exc:
    print(exc)
    return False
  
  return True
```

```python
is_write = write_file(
    file_path='func_ages.csv',
    header='age',
    content=ages
)
print(is_write)
```

```python
def calc_juros(initial_value, tax, years):
  final_value = initial_value

  for year in range(1, years + 1):
    final_value =+ final_value * (1 + tax)
  
  final_value = round(final_value, 2)
  return initial_value, tax, years, final_value

def print_calc(initial_value, tax, years, final_value):
  print(f'Valor inicial: R$ {initial_value}, taxa de juros anual: {tax} em {years} anos, valor final R$ {final_value}')

```

```python
initial_value, tax, years, final_value = calc_juros(1000.00, 0.05, 10)
print_calc(initial_value, tax, years, final_value)

initial_value, tax, years, final_value = calc_juros(1020.00, 0.03, 10)
print_calc(initial_value, tax, years, final_value)
```

## Escopo

```python
def list_sum(num: list) -> int:
  n = 0

  for number in num:
    n += number
  
  return n
```

```python
list_test = [2] * 20

soma = list_sum(num=list_test)
print(soma)
```
