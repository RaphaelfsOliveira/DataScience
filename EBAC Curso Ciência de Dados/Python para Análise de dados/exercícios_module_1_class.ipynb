# 3. Strings
> Aplique três métodos distintos na string abaixo, você pode conferir alguns métodos neste link:

```python
cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

m_split = cancao.split(',')
m_split = [c.strip() for c in m_split]
print(f'split with strip: {m_split}', end='\n\n')

m_find = cancao.find('moinho')
print(f'find -> moinho: {m_find}', end='\n\n')

m_upper = cancao.upper()
print(f'upper: {m_upper}', end='\n\n')

m_replace = m_upper.replace(' ', '').replace('RODA', ' ')
print(f'replace by upper: {m_replace}', end='\n\n')
```

> Extraia da string abaixo o valor da taxa selic na variável selic e o valor do ano na variavel
ano . Imprima os valores na tela.

```python
noticia = 'Selic vai a 2,75% e supera expectativas; ' + \
 'é a primeira alta em 6 anos.'

i_selic = noticia.find('2')
f_selic = noticia.find('%')

selic = float(noticia[i_selic:f_selic].replace(',', '.'))
print(f'Taxa selic: {selic}%')

year = noticia.find('anos')
year -= 2

ano = noticia[year]
print(f'Anos: {ano}')
```
