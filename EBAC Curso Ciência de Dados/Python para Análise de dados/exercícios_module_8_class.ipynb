```python
def valor_total_emprestimo(valor: float, quantidade: int) -> float:
  return valor * quantidade
```

```python
def get_data(linha_elementos):
  try:
    data = {
      'id_vendedor': linha_elementos[0],
      'valor_emprestimos': float(linha_elementos[1]),
      'quantidade_emprestimos': int(linha_elementos[2]),
      'data': linha_elementos[3],  
    }

  except ValueError as exc:
    data = {
      'id_vendedor': linha_elementos[0],
      'valor_emprestimos': float(linha_elementos[1].replace('"', '')),
      'quantidade_emprestimos': int(linha_elementos[2]),
      'data': linha_elementos[3],  
    }
    print(f'File error: {linha_elementos[0]} | error: {exc}')
  
  return data
```

```python
def read_file(file_name):
  emprestimos = []

  with open(file=file_name, mode='r', encoding='utf8') as fp:
    fp.readline() # cabeçalho
  
    linha = fp.readline()
    while linha:
      try:
        linha_emprestimo = {}
        linha_elementos = linha.strip().split(sep=',')
        linha_emprestimo = get_data(linha_elementos)
      
      except Exception as exc:
        print(f'{exc}')

      finally:
        emprestimos.append(linha_emprestimo)
        linha = fp.readline()
  
  return emprestimos
```

```python
def get_values(emprestimos: list, valor_total_emprestimo):
  emprestimos_total = []

  for emprestimo in emprestimos:
    valor_total = valor_total_emprestimo(
        valor=emprestimo['valor_emprestimos'],
        quantidade=emprestimo['quantidade_emprestimos']
    )
    emprestimos_total.append({emprestimo['id_vendedor']: valor_total})
  
  return emprestimos_total
```

```python
emprestimos = read_file(file_name='./credito.csv')
emprestimos_total = get_values(emprestimos, valor_total_emprestimo)

for emprestimo_total in emprestimos_total:
  print(emprestimo_total)
```
