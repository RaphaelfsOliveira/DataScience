# Python: Programação Funcional

## Função lambda

```python
extract_email_provider = lambda e: e.split(sep='@')[-1]
```

```python
email = 'raphael@email.com'
provider = extract_email_provider(email)

print(provider)
```

```python
num_par = lambda num: True if num % 2 == 0 else False
```

```python
numbers = range(0, 10)

for num in numbers:
  if num_par(num):
    print(f'O número {num} é par')
```

```python
def retorno(juros: float):
  return lambda invest: invest * (1 + juros)
```

```python
retorno_5 = retorno(juros=0.05)
retorno_10 = retorno(juros=0.10)
```

```python
final_value = retorno_5(invest=1000)
print(final_value)

final_value = retorno_10(invest=1000)
print(final_value)
```

```python
years = 10
initial_value = 1000
final_value = initial_value

for year in range(1, years + 1):
  final_value = retorno_5(invest=final_value)

final_value = round(final_value, 2)
print(final_value)
```

```python
years = 10
initial_value = 1000
final_value = initial_value

for year in range(1, years + 1):
  final_value = retorno_10(invest=final_value)

final_value = round(final_value, 2)
print(final_value)
```

## Função map

```python
nums = [1, 2, 3]

num_cube = map(lambda num: num ** 3, nums)

print(list(num_cube))
```

```python
emails = ['email@gmail.com', 'email@msn.com', 'email@orkut.com']
extract_provider = lambda email: email.split(sep='@')[-1]
```

```python
providers = []

for email in emails:
  provider = extract_provider(email)
  providers.append(provider)

print(providers)
```

```python
providers = list(map(extract_provider, emails))
print(providers)
```

```python
years = [10, 10, 10]
tax = [0.05, 0.1, 0.15]
initial_value = [1000, 1000, 1000]

def retorno(initial_value: float, tax: float, years: int) -> float:
  final_value = initial_value

  for year in range(1, years + 1):
    final_value *= (1 + tax)
  
  return round(final_value, 2)

test = list(map(retorno, initial_value, tax, years))
print(test)
```

## Função filter

```python
numbers = [1, 2, 3]

impar = filter(lambda num: num % 2 != 0, numbers)

print(list(impar))
```

```python
emails = ['raphael@email.com', 'raphael@orkut.com', 'raphael@msn.com']
provider_google = filter(lambda email: 'orkut' in email, emails)

only_emails = filter(lambda email: '@' in email, emails)

print(list(provider_google))
print(list(only_emails))
```

## Função reduce

```python
from functools import reduce

numbers = [1, 2, 3]
numbers = [1, 2, 3, 4, 5]

soma = reduce(lambda x, y: x + y, numbers)
print(soma)
```

```python
from functools import reduce
from random import random

def max_then(first: int, second: int) -> int:
  return first if first >= second else second

first = 1
second = 10

print(reduce(max_then, [first, second]))
```

```python
numbers = [round(100 * random()) for _ in range(0, 100)]
print(numbers)
```

```python
max_number = reduce(max_then, numbers)
print(max_number)
```

```python
max_number = reduce(lambda first, second: first if first >= second else second, numbers)
print(max_number)
```
