> 1.1. Crie uma lista chamada paises com o pais dos usuários com mais de 30 anos.

> 1.3. Crie uma lista chamada cidades com a cidade dos usuários que passaram mais de 70
segundos no site.

```python
cities = []
percentage_cities = []


def site_time(data):
  site_time = 70
  stayed_site = None
  time = None

  try:
    stayed_site = data['tempo_gasto_site'] > site_time
    time = data['tempo_gasto_site']

  except TypeError:
    time = data['tempo_gasto_site']
    
    if time:
      time = float(data['tempo_gasto_site'])
      stayed_site = time > site_time

  except Exception as error:
    print(f'{error}')

  return stayed_site, time


def internet_time(data):
  time = None

  try:
    time = data['tempo_gasto_internet']
  
  except TypeError:
    time = data['tempo_gasto_internet']
    if time:
      time = float(internet_time)

  except Exception as error:
    print(f'{error}')
  
  return time


def percentage_site(stay_site, internet_time):
    percentage = None

    calc_site = isinstance(stay_site, float)
    calc_internet = isinstance(internet_time, float)
    active_user = (calc_site and calc_internet)

    if active_user:
      percentage = round((stay_site * 100) / internet_time, 2)
  
    return percentage


for data in propaganda_online:
  stayed_site, time_in_site = site_time(data)
  time = internet_time(data)
  percentage = percentage_site(time_in_site, time)
  
  if stayed_site:
    city = data['cidade']
    cities.append(city)

    percentage_cities.append({
        'city': city,
        'percent': f'{percentage} %',
        'time_in_site': f'{time_in_site} seconds',
    })

temp_list = list(set(cities))
print(f"Cidades, mais de 70 seg no site: {', '.join(temp_list)}", end='\n\n')

print('Percentage of time on site by city', end='\n\n')
for data in percentage_cities:
  for key, value in data.items():
    print(f'{key}: {value}')
  
  print('\n')
```
