## 1. Funções para arquivo csv

```python
def data_type(data, type_):
  types = {
      'str': str,
      'int': int,
      'float': float
  }

  try:
    return types[type_](data)
  
  except TypeError as exc:
    print(f'{exc}')
    return False
  
  except Exception as exc:
    print(f'{exc}')
    return False
  
  return data


def extrai_coluna_csv(nome_arquivo: str, indice_coluna: int, tipo_dado: str, sep: str = ','):
  coluna = []

  with open(file=nome_arquivo, mode='r', encoding='utf8') as fp:
    line = fp.readline() # header
    line = fp.readline()

    while line:
      colunas = line.split(sep)
      data = colunas[indice_coluna]
      data = data_type(data, tipo_dado)

      coluna.append(data)
      
      line = fp.readline()

  return coluna
```

```python
# extrair a coluna valor_venda
valor_venda = extrai_coluna_csv(
    nome_arquivo='./carros.csv',
    indice_coluna=1,
    tipo_dado='str'
)
print(valor_venda) # deve retornar ['vhigh', 'med', 'low', ...]
```

## 2. Funções para arquivo txt

```python
def extrai_linha_txt(nome_arquivo: str, numero_linha: int, sep: str = ' '):
  palavras_linha = []
  
  with open(file=nome_arquivo, mode='r', encoding='utf8') as fp:
    content = fp.readlines()
    
    line = content[numero_linha - 2]
    line = line.replace('\n', '')
    
    words = line.split(sep)
    palavras_linha = words

  return palavras_linha
```

```python
linha10 = extrai_linha_txt(
    nome_arquivo='./musica.txt',
    numero_linha=10
)
print(linha10)
# deve retornar ['Mas', 'eis', 'que', 'chega', 'a', 'roda', 'viva']
```
